### PACKAGES IMPORT

```python
import os
import uuid
import datetime 
import pandas as pd
from icalendar import Calendar, Event, vDatetime  
```

### CALENDAR IMPORT

```python
group_name = input("Enter group name: ")
```

```python
classes = pd.read_csv(f"schedules/{group_name}.csv", encoding="UTF-8")

classes
```

```python
if not os.path.exists("calendars"):
    os.mkdir("calendars")
```

```python
calendar = Calendar()
calendar["dtstart"] = "20260223T000000"
calendar["dtend"] = "20260930T235999"
calendar["summary"] = f"Schedule of the {group_name} group."
calendar["version"] = "2.0"
calendar["PRODID"] = f"-//Apollo schedule export//{group_name}"
calendar["tzname"] = "CEST"
```

```python
for index, row in classes.iterrows():
    event = Event()
    event["dtstart"] = row["Termin"].replace("-", "") + "T" + row["Start time"].replace(":", "") + "00"
    event["dtend"] = row["Termin"].replace("-", "") + "T" + row["End time"].replace(":", "") + "00"
    event["location"] = row["Sala"]
    event["summary"] = row["Przedmiot"]
    event["description"] = row["Nauczyciel"]
    event["UID"] = uuid.uuid4()
    event["DTSTAMP"] = vDatetime(datetime.datetime.now())
    calendar.add_component(event)
```

```python
cf = open(f"calendars/{group_name}.ics", "w", encoding="UTF-8")
cf.write(calendar.to_ical().decode("UTF-8"))
cf.close()
```
